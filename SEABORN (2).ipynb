```python
#SEABORN >> Seaborn is a Python data visualization library based on matplotlib.
#It provides a high-level interface for drawing attractive and informative statistical graphics.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warning
warning.filterwarning('ignore')
```

```python
# numerical data plotting
#relplot --> relationship between categorial data and numerical data or relationship between two variables.
#relplot --> relationship plot
```

```python

df
```

```python
#how pulse is increasing with time >> scatterplot / bydefault relplot makes scatter plot

sns.relplot (x = 'pulse', y = 'time' , data = df) # relplot showing relationship between two variables
```

```python
sns.relplot (x = 'id', y = 'pulse' , data = df , hue = 'diet') # hue differentiates by using different color shades
```

```python

sns.relplot (x = 'id', y = 'pulse' , data = df , hue = 'kind' )
```

```python
time = np.arange(100)
value = np.random.rand(100)
df = pd.DataFrame({'time': time , 'value': value})
df
```

```python
#line plot
sns.relplot(x = 'time', y = 'value', kind='line', data = df , sort=True)
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = 'timepoint', y='signal', kind='line', data=fmri, hue='event' , style='event')
```

```python
#removes dashes from the chart
sns.relplot(x = 'timepoint', y='signal', kind='line', data=fmri, hue='event' , style='event', dashes = False)
```

```python
sns.relplot(x = 'timepoint', y='signal', kind='line', data=fmri, hue='event' , style='event',markers=True)
```
